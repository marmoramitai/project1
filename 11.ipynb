# Income male vs female #

```python
sorted_data = (
    after_cuting_df.groupby(['Household Head Marital Status','Household Head Sex'])['Total Household Income']
    .mean()
    .reset_index()
)

plt.figure(figsize=(12, 6))
sns.barplot(
    data=sorted_data,
    x='Household Head Marital Status',
    y='Total Household Income',
    hue='Household Head Sex',
    palette=category_colors
)

plt.title('Total Household Income by Sex and Marital Status')
plt.ylabel('Total Household Income')
plt.xlabel('Household Head Sex')

plt.show()
```

### The total household income is significantly higher when the 'Household Head Sex' is female, regardless of the 'Household Head Marital Status' ###

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(16, 12))

categories = ['Married', 'Single', 'Divorced/Separated', 'Widowed']
category_titles = ['Married Household Heads', 'Single Household Heads', 'Divorced Household Heads', 'Widowed Household Heads']

for i, category in enumerate(categories):
    row = i // 2
    col = i % 2
    
    sns.stripplot(
        data=after_cuting_df[after_cuting_df['Household Head Marital Status'] == category], 
        y='Education Expenditure', 
        hue='Household Head Sex',  
        dodge=True,
        jitter=0.3,
        alpha=0.7,
        ax=ax[row, col],
        size=8,
        palette=category_colors
    )
    ax[row, col].set_title(category_titles[i])
    ax[row, col].set_ylabel('Education Expenditure')
    ax[row, col].set_xlabel('Household Head Sex')

plt.tight_layout()
plt.show()
```

## This analysis helps in understanding how marital status and the sex of the household head might influence the distribution of education-related spending ##

```python
order = after_cuting_df.groupby(['Household Head Marital Status',"Household Head Sex"])[['Education Expenditure']].mean().reset_index()

plt.figure(figsize=(20, 8))

sns.barplot(
    data=after_cuting_df, 
    x='Household Head Marital Status', 
    y='Education Expenditure', 
    hue='Household Head Sex',
    palette=category_colors,
    errorbar=None,
)

plt.title('Average Education Expenditure by Gender and Marital Status', fontsize=24)
plt.ylabel('Average Education Expenditure', fontsize=18)
plt.xlabel('', fontsize=18)

plt.xticks(fontsize=18)
plt.legend(title='Household Head Sex', fontsize=18, title_fontsize=20)

plt.tight_layout()

plt.show()
```

## As we can see, no matter what's your marital status, female spend more money on education than male ##

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(14, 8))
plt.subplots_adjust(hspace=0.4, wspace=0.4)

sns.barplot(data=after_cuting_df[after_cuting_df['Household Head Marital Status'] == 'Divorced/Separated'], 
            y='Education Expenditure', x='Members with age 5 - 17 years old', 
            hue='Household Head Sex', ax=ax[0, 0], palette=category_colors, errorbar=None)
ax[0, 0].set_title('Divorced/Separated', loc='center')

sns.barplot(data=after_cuting_df[after_cuting_df['Household Head Marital Status'] == 'Married'], 
            y='Education Expenditure', x='Members with age 5 - 17 years old', 
            hue='Household Head Sex', ax=ax[0, 1], palette=category_colors, errorbar=None)
ax[0, 1].set_title('Married', loc='center')

sns.barplot(data=after_cuting_df[after_cuting_df['Household Head Marital Status'] == 'Widowed'], 
            y='Education Expenditure', x='Members with age 5 - 17 years old', 
            hue='Household Head Sex', ax=ax[1, 0], palette=category_colors, errorbar=None)
ax[1, 0].set_title('Widowed', loc='center')

sns.barplot(data=after_cuting_df[after_cuting_df['Household Head Marital Status'] == 'Single'], 
            y='Education Expenditure', x='Members with age 5 - 17 years old', 
            hue='Household Head Sex', ax=ax[1, 1], palette=category_colors, errorbar=None)
ax[1, 1].set_title('Single', loc='center')

for i in range(2):
    for j in range(2):
        ax[i, j].set_xlabel('Members Age 5-17', fontsize=14)
        ax[i, j].set_ylabel('Education Expenditure', fontsize=14)
        ax[i, j].set_ylim(0, 25000)

plt.suptitle('Average Education Expenditure by Gender, Marital Status and number of children', fontsize=20, y=1.02)
plt.show()
```

#### We have to check the percents of the invest in education both income and outcome ####

```python
after_cuting_df = after_cuting_df.copy()

after_cuting_df['outcome'] = (
    after_cuting_df['Bread and Cereals Expenditure'] + 
    after_cuting_df['Total Rice Expenditure'] +
    after_cuting_df['Meat Expenditure'] +
    after_cuting_df['Total Fish and  marine products Expenditure'] +
    after_cuting_df['Fruit Expenditure'] +
    after_cuting_df['Vegetables Expenditure'] +
    after_cuting_df['Restaurant and hotels Expenditure'] +
    after_cuting_df['Alcoholic Beverages Expenditure'] +
    after_cuting_df['Tobacco Expenditure'] +
    after_cuting_df['Clothing, Footwear and Other Wear Expenditure'] +
    after_cuting_df['Housing and water Expenditure'] +
    after_cuting_df['Imputed House Rental Value'] +
    after_cuting_df['Medical Care Expenditure'] +
    after_cuting_df['Transportation Expenditure'] +
    after_cuting_df['Communication Expenditure'] +
    after_cuting_df['Education Expenditure'] +
    after_cuting_df['Miscellaneous Goods and Services Expenditure'] +
    after_cuting_df['Special Occasions Expenditure'] +
    after_cuting_df['Crop Farming and Gardening expenses']
)

after_cuting_df['education per income'] = (
    after_cuting_df['Education Expenditure'] / after_cuting_df['Total Household Income']
)

after_cuting_df['education per outcome'] = (
    after_cuting_df['Education Expenditure'] / after_cuting_df['outcome']
)
```

```python
after_cuting_df.head()
```

### The column 'outcome' sums up the various expenditure categories in one step, directly computing the total household expenditure. ###
### education per income: This column represents the proportion of a household's income that is being spent on education  ###
### education per outcome: This column represents the proportion of overall household expenditure spent on education ###

```python
fig, ax = plt.subplots(nrows=2, figsize=(14, 8), ncols=2)
sns.barplot(data=after_cuting_df[after_cuting_df['Household Head Marital Status'] == 'Divorced/Separated'], y = 'education per income', x = 'Members with age 5 - 17 years old', hue = 'Household Head Sex',
            ax=ax[0, 0], palette = category_colors,  errorbar=None
)
sns.barplot(data=after_cuting_df[after_cuting_df['Household Head Marital Status'] == 'Married'], y = 'education per income', x = 'Members with age 5 - 17 years old', hue = 'Household Head Sex', 
            ax=ax[0, 1], palette = category_colors, errorbar=None

)
sns.barplot(data=after_cuting_df[after_cuting_df['Household Head Marital Status'] == 'Widowed'], y = 'education per income', x = 'Members with age 5 - 17 years old', hue = 'Household Head Sex',
            ax=ax[1, 0], palette = category_colors, errorbar=None

)
sns.barplot(data=after_cuting_df[after_cuting_df['Household Head Marital Status'] == 'Single'], y = 'education per income', x = 'Members with age 5 - 17 years old', hue = 'Household Head Sex', 
            ax=ax[1, 1], palette = category_colors, errorbar=None

)
ax[0, 0].set_title('Divorced/Separated', loc='center', fontsize=16)
ax[0, 1].set_title('Married', loc='center', fontsize=16)
ax[1, 0].set_title('Widowed', loc='center', fontsize=16)
ax[1, 1].set_title('Single', loc='center', fontsize=16)

ax[0, 0].set_ylim(0, 0.1)
ax[0, 1].set_ylim(0, 0.1)
ax[1, 0].set_ylim(0, 0.1)
ax[1, 1].set_ylim(0, 0.1)
plt.suptitle('Average Education Expenditure per income by Gender, Marital Status and children', fontsize=20, y=1.02)
for i in range(2):
    for j in range(2):
        ax[i, j].set_xlabel('Members Age 5-17', fontsize=14)
        ax[i, j].set_ylabel('Education per income', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
ncr_data = {
    'Main Source of Income': ['Wage', 'Entrepreneurial Activities'],
    'Income ≥ 0': [len(ncr_df[(ncr_df['Main Source of Income'] == 'Wage/Salaries') & (ncr_df['income less outcome'] >= 0)]), len(ncr_df[(ncr_df['Main Source of Income'] == 'Enterpreneurial Activities') & (ncr_df['income less outcome'] >= 0)])],
    'Income < 0': [len(ncr_df[(ncr_df['Main Source of Income'] == 'Wage/Salaries') & (ncr_df['income less outcome'] < 0)]), len(ncr_df[(ncr_df['Main Source of Income'] == 'Enterpreneurial Activities') & (ncr_df['income less outcome'] < 0)])],
}

categories_ncr = ncr_data['Main Source of Income']
income_positive_ncr = np.array(ncr_data['Income ≥ 0'])
income_negative_ncr = np.array(ncr_data['Income < 0'])

total_positive_ncr = sum(income_positive_ncr)
total_negative_ncr = sum(income_negative_ncr)

income_positive_percent_ncr = income_positive_ncr / total_positive_ncr * 100
income_negative_percent_ncr = income_negative_ncr / total_negative_ncr * 100

barmm_data = {
    'Main Source of Income': ['Wage', 'Entrepreneurial Activities'],
    'Income ≥ 0': [len(barmm_df[(barmm_df['Main Source of Income'] == 'Wage/Salaries') & (barmm_df['income less outcome'] >= 0)]), len(barmm_df[(barmm_df['Main Source of Income'] == 'Enterpreneurial Activities') & (barmm_df['income less outcome'] >= 0)])],
    'Income < 0': [len(barmm_df[(barmm_df['Main Source of Income'] == 'Wage/Salaries') & (barmm_df['income less outcome'] < 0)]), len(barmm_df[(barmm_df['Main Source of Income'] == 'Enterpreneurial Activities') & (barmm_df['income less outcome'] < 0)])],
}

categories_barmm = barmm_data['Main Source of Income']
income_positive_barmm = np.array(barmm_data['Income ≥ 0'])
income_negative_barmm = np.array(barmm_data['Income < 0'])

total_positive_barmm = sum(income_positive_barmm)
total_negative_barmm = sum(income_negative_barmm)

income_positive_percent_barmm = income_positive_barmm / total_positive_barmm * 100
income_negative_percent_barmm = income_negative_barmm / total_negative_barmm * 100

fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# Plot NCR data
bar_width = 0.35
x_ncr = np.arange(len(categories_ncr))

axes[0].bar(x_ncr - bar_width / 2, income_positive_percent_ncr, width=bar_width, label='Income ≥ 0 (Positive)', color='gold')
axes[0].bar(x_ncr + bar_width / 2, income_negative_percent_ncr, width=bar_width, label='Income < 0 (Negative)', color='brown')

axes[0].set_xticks(x_ncr)
axes[0].set_xticklabels(categories_ncr)
axes[0].set_ylabel('Percentage', fontsize=14)
axes[0].set_title('Income Source Distribution by Income Level in NCR', fontsize=16)
axes[0].legend(loc='best')

for i, v in enumerate(income_positive_percent_ncr):
    axes[0].text(i - bar_width / 2, v + 1, f"{v:.1f}%", ha='center', color='black')
for i, v in enumerate(income_negative_percent_ncr):
    axes[0].text(i + bar_width / 2, v + 1, f"{v:.1f}%", ha='center', color='black')

# Plot BARMM data
x_barmm = np.arange(len(categories_barmm))

axes[1].bar(x_barmm - bar_width / 2, income_positive_percent_barmm, width=bar_width, label='Income ≥ 0 (Positive)', color='gold')
axes[1].bar(x_barmm + bar_width / 2, income_negative_percent_barmm, width=bar_width, label='Income < 0 (Negative)', color='brown')

axes[1].set_xticks(x_barmm)
axes[1].set_xticklabels(categories_barmm)
axes[1].set_ylabel('Percentage', fontsize=14)
axes[1].set_title('Income Source Distribution by Income Level in BARMM', fontsize=16)
axes[1].legend(loc='best')

for i, v in enumerate(income_positive_percent_barmm):
    axes[1].text(i - bar_width / 2, v + 1, f"{v:.1f}%", ha='center', color='black')
for i, v in enumerate(income_negative_percent_barmm):
    axes[1].text(i + bar_width / 2, v + 1, f"{v:.1f}%", ha='center', color='black')

plt.tight_layout()
plt.show()
```
